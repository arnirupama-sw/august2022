```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://s3.eu-west-1.amazonaws.com/neueda.conygre.com/pydata/SPX.csv',
                     index_col='Date', parse_dates=True)
```

```python
df.dtypes
```

```python
df.head()
```

```python
df['Price'] = pd.to_numeric(df['Price'].str.replace(',',''))
df['Open'] = pd.to_numeric(df['Open'].str.replace(',',''))
df['High'] = pd.to_numeric(df['High'].str.replace(',',''))
df['Low'] = pd.to_numeric(df['Low'].str.replace(',',''))
df['Change %'] = pd.to_numeric(df['Change %'].str.replace('%',''))
```

```python
df.dtypes
```

```python
df.head()
```

```python
df = df.sort_values('Date')
df
```

```python
df
```

```python
data = df.query("Date >= 2015 and Date < 2017")[['High','Low']]
```

```python
data.plot(figsize=(18,9))
```

```python
by_BMonth = pd.Grouper(freq='BM')
funcs=['mean','min','max']
data2 = df.groupby(by_BMonth).agg(funcs)
data2
```

```python
data2.plot(figsize=(15,15))
```
